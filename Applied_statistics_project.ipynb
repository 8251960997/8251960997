Distribution of Income with fitted Distribution

```python
from scipy.stats import norm

# Plot the distribution of 'Income'
plt.figure(figsize=(10, 6))
sns.histplot(df['Income'], kde=True, color='skyblue', label='Income Distribution')

mu, std = norm.fit(df['Income'])
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2, label='Fitted Normal Distribution')

plt.title('Distribution of Income with Fitted Normal Distribution')
plt.xlabel('Income')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Probability of Successes in Binomial Distribution

```python
n = 10  # number of trials
p = 0.5  # probability of success
successes = np.arange(0, n+1)
probabilities = binom.pmf(successes, n, p)

plt.figure(figsize=(10, 6))
plt.bar(successes, probabilities, color='skyblue')
plt.title('Probability Distribution for Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(successes)
plt.show()
```

```python
# Example: Select 'Income' and 'NumDealsPurchases' for segmentation
X = df[['Income', 'NumDealsPurchases']]
```

```python
# Example: Perform K-means clustering
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
df['Cluster'] = kmeans.fit_predict(X)
```
